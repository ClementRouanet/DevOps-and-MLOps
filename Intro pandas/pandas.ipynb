```python
import pandas as pd
```

```python
# Exercice 0

# Importez les données SeaCareerSet
fichier = "SeaCareerSet.csv"
SeaCareerSet = pd.read_csv(fichier)

SeaCareerSet.head()
```

```python
# Exercice 1

# 1. Utilisez la méthode adéquate afin d’afficher les 10 premières valeurs
print(SeaCareerSet.head(10))


# 2. Utilisez la méthode adéquate afin d’afficher les 15 dernières
print(SeaCareerSet.tail(15))


# 3. Utilisez la méthode adéquate afin d’afficher un échantillon aléatoire de 20 valeurs
echantillon = SeaCareerSet.sample(20)
print(echantillon)


# 4. Exportez l’échantillon dans un nouveau fichier CSV
fichier = "echantillon.csv"
echantillon.to_csv(fichier, index=False)
```

```python
# Exercice 2

# Affichez le nombre de lignes et de colonnes en une seule commande python
print("Nombre de lignes et de colonnes :", SeaCareerSet.shape)


# Obtenez des statistiques descriptives pour les colonnes numériques
print(SeaCareerSet.describe())


# Obtenez des statistiques descriptives pour les colonnes non numériques
print(SeaCareerSet.describe(include='object'))


# Affichez les valeurs uniques de la colonne contenant les catégories de travail. Que constatez vous ?
print(SeaCareerSet['workclass'].unique())
    # Il y a des valeurs 'nan' et '?'
```

```python
# Exercice 3

# Filtrez les données pour n'afficher que les passagers ayant obtenu une Maîtrise
passagers_maitrise = SeaCareerSet[SeaCareerSet['education'] == 'Masters']


# Filtrez les données pour n'afficher que les passagers majeurs
passagers_majeurs = SeaCareerSet[SeaCareerSet['age'] >= 18]


# Créez un DataFrame SinglePassengers avec uniquement les passagers voyageant seul
SinglePassengers = SeaCareerSet[(SeaCareerSet['sibsp'] == 0) & (SeaCareerSet['parch'] == 0)]


# Sélectionnez dans SinglePassengers les colonnes avec des informations que l’on peux retrouver sur un passeport
print(SinglePassengers[['name', 'sex', 'age', 'native_country']])
```

```python
# Exercice 4

# Identifiez les valeurs manquantes dans le DataFrame SeaCareerSet
valeurs_manquantes = SeaCareerSet.isnull().sum()
print(valeurs_manquantes)


# Grâce a une information dans une autre colonne, une colonne incomplète peut être compléter. Identifiez et complétez cette colonne


# Remplacez les valeurs de chaque colonne en utilisant au minimum 3 méthode différente
mean_hpw = SeaCareerSet['hours_per_week'].mean()
SeaCareerSet['hours_per_week'].fillna(mean_hpw, inplace=True)

SeaCareerSet['education'].fillna('Unknown', inplace=True)
SeaCareerSet['marital_status'].fillna('Unknown', inplace=True)
SeaCareerSet['relationship'].fillna('Unknown', inplace=True)
SeaCareerSet['embarked'].fillna('Unknown', inplace=True)
SeaCareerSet['workclass'].fillna('Unknown', inplace=True)
SeaCareerSet['occupation'].fillna('Unknown', inplace=True)
SeaCareerSet['native_country'].fillna('Unknown', inplace=True)
```

```python
# Exercice 5

# Triez les données en fonction du nombre d’heures de travail par semaine. Du plus travailleurs au moins travailleurs
donnees_triees = SeaCareerSet.sort_values(by='hours_per_week', ascending=False)


# Calculez la moyenne des dettes de chaque profession
moyenne_dettes_par_profession = SeaCareerSet.groupby('occupation')['capital_loss'].mean()


# Calculez le patrimoine médian de chaque profession
patrimoine_median_par_profession = SeaCareerSet.groupby('occupation')['capital_gain'].median()


# Calculez le min et le max de chaque colonne en groupant les données par tranche de salaire
min_max_par_salaire = SeaCareerSet.groupby('income').agg({'min', 'max'})
```

```python
# Exercice 6

# Créez un histogramme de l'âge des passagers avec titre et noms de axes
plt = SeaCareerSet['age'].hist(bins=20, color='skyblue', edgecolor='black')

plt.set_title("Répartition de lâge des passagers")
plt.set_xlabel('Âge')
plt.set_ylabel('Nombre de passagers')
```
